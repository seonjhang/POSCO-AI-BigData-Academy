## 성적 관리 프로그램

#### (문제)

파일로부터 데이터를 읽어서 성적 목록을 만들어 관리하는 성적 관리 프로그램을 작성한다.

```python
import sys
arg = sys.argv
if len(arg) > 6:
    a,b = arg.split()
    filename = str(b)
    infile = open(filename, "r")
    
else:
    infile = open('students.txt', 'r')

#성 + 이름 합치기
stu_list = []

for line in infile:
    L = line.split() 
    revised = (L[1] + ' ' + L[2])  
    new_L = [L[0], revised, L[3], L[4]]

    stu_list += [new_L]

i = 0
for i in range(len(stu_list)):
    a = (int(stu_list[i][2]) + int(stu_list[i][3]))/2
    stu_list[i].append(a)
    i += 1

def average(m, f):
    return (m + f) / 2

def grade(avg):
    if avg >= 90:
        grade = "A"
    elif 80 <= avg < 90:
        grade = "B"
    elif 70 <= avg < 80:
        grade = "C"
    elif 60 <= avg < 70:
        grade = "D"
    else:
        grade = "F"

    return grade


i = 0
for i in range(len(stu_list)):
    b = grade(stu_list[i][4])
    stu_list[i].append(b)
    i += 1

stu_list.sort(key=lambda e : e[4], reverse = True)

# else:
#     print("%s file dese not exists\n" % filename)  
    
while True:
    _input = input("# ").lower()
    if _input == 'show':
        show(stu_list)
    elif _input == 'search':
        show(stu_list)
    elif _input == 'changescore':
        show(stu_list)
    elif _input == 'add':
        show(stu_list)
    elif _input == 'searchgrade':
        show(stu_list)
    elif _input == 'remove':
        show(stu_list)
    elif input == 'quit':
        infile.close()
        quit(stu_list)
        break
    else:
        continue

    
    
# show

def show(stu_list):
    show = input("# ")

    print("  Student   \t Name       Midterm    Final   Average Grade")
    print('-'*65)
    stu_list.sort(key=lambda e : e[4], reverse = True)

    for stu in stu_list:
        temp = str(stu[0])+'\t'+str(stu[1])+'\t'+str(stu[2])+'\t'+str(stu[3])+'\t'+str(stu[4])+'\t'+str(stu[5])+'\n'
        print(temp)    
    print()
    print('#')
    
    return



# search

def search(stu_list):
    search = input("# ")
    s_id = input("Student ID: ")
    
    i = 0

    for i in range(len(stu_list)):
        if stu_list[i][0] == s_id:
            print("  Student   \t Name       Midterm    Final   Average Grade")
            print('-'*65)
            for stu in stu_list:
                temp = str(stu[0])+'\t'+str(stu[1])+'\t'+str(stu[2])+'\t'+str(stu[3])+'\t'+str(stu[4])+'\t'+str(stu[5])+'\n'
                print(temp)    
            i += 1
        
            return
  
    print("NO SUCH PERSON.")

    print()
    print('#')
    return



#changescore (점수 수정)

def changescore(stu_list):
    
    change_score = input("# ")
    s_id = input("Student ID: ")
    
    i = 0
    for i in range(len(stu_list)):
        if stu_list[i][0] == s_id:
            score = input("Mid/Final? ")
           
            if score == "mid":
                n_score = input("input new score: ")
                if int(n_score) <= 100:
                    a = int(n_score)
                    avg = average(a, int(stu_list[i][3]))
                    g =  str(grade(avg))

                    stu_list[i][2] = a
                    stu_list[i][4] = avg
                    stu_list[i][5] = g
                    
                    print("  Student   \t Name       Midterm    Final   Average Grade")
                    print('-'*65)
                    
                    for stu in stu_list:
                        temp = str(stu[0])+'\t'+str(stu[1])+'\t'+str(stu[2])+'\t'+str(stu[3])+'\t'+str(stu[4])+'\t'+str(stu[5])+'\n'
                        print(temp)                       

                else:
                    print("0~100 사이의 숫자를 입력하시오.")
                
                return

            elif score == "final":
                n_score = input("input new score: ")
                if int(n_score) <= 100:
                    b = int(n_score)
                    avg = average(int(stu_list[i][3]), b)
                    g = str(grade(avg))

                    stu_list[i][3] = b
                    stu_list[i][4] = avg
                    stu_list[i][5] = g
                    
                    print("  Student   \t Name       Midterm    Final   Average Grade")
                    print('-'*65)
                    
                    for stu in stu_list:
                        temp = str(stu[0])+'\t'+str(stu[1])+'\t'+str(stu[2])+'\t'+str(stu[3])+'\t'+str(stu[4])+'\t'+str(stu[5])+'\n'
                        print(temp)                       

                    return
                
            else:
                print("오타를 확인하시오.")
                
                return
            
            
    
    
    print("NO SUCH PERSON.")
    


# add (학생 추가)

def add(stu_list):
    
    add = input("# ")
    s_id = input("Student ID: ")
    
    i = 0
    for i in range(len(stu_list)):
        if stu_list[i][0] == s_id:
            i += 1
            print("ALREADY EXISTS.")
            return

    name = input("Name: ")
    m = int(input("Midterm Score: "))
    f = int(input("Final Score: "))
    a  = [s_id, name, m, f, average(m,f), grade(average(m,f))]
    stu_list.extend([a])
    print("Student added.")



# searchgrade (Grade 검색)

def search_g(stu_list):
    search_g = input("Grade to search: ")

    if search_g not in ['A', 'B', 'C', 'D', 'F']:
        print("오타를 확인하시오.")
        
        return
    
    result = False
    for stu in stu_list:
        if stu[5] == search_g:
            result = True
            print(stu)
            
    if result == False:        
        print("NO RESULTS.")
            

# REMOVE (특정 학생 삭제)

def remove(stu_list):
    
    remove = input("# ")
    s_id = input("Student ID: ")
    
    if not stu_list:
        print("List is empty.")

    for stu in stu_list:
        if stu[0] == s_id:
            stu_list.remove(stu)
            print("Student removed.")
            return
        
    print("NO SUCH PERSON.")


# quit (종료)

def quit(stu_list):

    quit = input("# ")
    save = input("Save data? [yes/no]")

    if save == 'yes':
        filename = input("File name: ")
        fw = open(filename, "w")
        stu_list.sort(key=lambda e : e[4], reverse = True)

        for stu in stu_list:
            temp = str(stu[0])+'\t'+str(stu[1])+'\t'+str(stu[2])+'\t'+str(stu[3])+'\t'+str(stu[4])+'\t'+str(stu[5])+'\n'
            fw.write(temp)
        
        fw.close()

    elif save == 'no':
        return
```
